```python
person = {'name' : '홍길동', 'age' : 10}
# 요소 추가
person['gender'] = '남자'
print(person)
# 요소 수정
person['age'] = 20
print(person)
# 요소 삭제
del person['gender']
print(person)
# 함수, 메소드 설명보기 => help함수 이용
help(person.clear)
```

```python
# 딕셔너리에서 키 값 가져오기
print(person.keys())

# 타입 확인하기 내장함수 type() 사용
print(type(person.keys()))

print('=' * 20)

# 딕셔너리에서 값만 가져오기
print(person.values())
print(type(person.values()))

print('=' * 20)

# 딕셔너리에서 아이템(키, 값 쌍으로) 가져오기
print(person.items())
print(type(person.items()))

print('=' * 20)

# 딕셔너리의 key, values, 아이템 순회하기
for k in person.keys() :
  print(k)
print('=' * 20)
for v in person.values() :
  print(v)
print('=' * 20)
for i in person.items() :
  print(i)
print('=' * 20)

# 아이템 순회하기 (개별변수로 받아내기)
for k, v in person.items():
  print(f'{k} - {v}')

# 튜플 개별 변수에 저장하기
name, age, gender = ('홍길동', 30, '남자')
print(name, age, gender)
```

```python
# 집합자료형 Set(세트) 연습
s1 = { 1, 2, 3 }
s2 = set( [3, 4, 5] )
print (s1, s2)
print('=' * 20)

# 교집합
intersection = s1 & s2
print(intersection)
print('=' * 20)

# 차집합
difference = s1 - s2
print(difference)
print('=' * 20)

# 합집합
union = s1 | s2
print(f'합집합 {union}')
```

```python
def calculate(x, y) :
  add = x + y
  difference = x - y
  multifly = x * y
  mod = x // y
  remain = x % y
  return add, difference, multifly, mod, remain

print(type(calculate(10, 3)))
print(calculate(10, 3))

add, difference, multifly, mod, remain = calculate(10, 3)
print(add, difference, multifly, mod, remain)
```

```python
# for문의 range를 이용하여 1부터 100까지 합을 구하기
sum = 0
for num in range(1, 100+1) :
  sum += num
print(f'sum={sum}')
```

```python
# 문제) 1~100까지 정수 중에 짝수의 합과 홀수의 합을 각각 구하시오

sum_even = 0 ; sum_odd = 0

for num in range(1, 100+1) :
  if (num % 2 == 0) :
    sum_even += num
  else :
    sum_odd += num

print(f'짝수의 합 : {sum_even}')
print(f'홀수의 합 : {sum_odd}')
```

```python
# 문제) 구구단 출력하기 2~9단까지 중첩 for문 이용

for i in range(2, 10) :
  print(f'====={i}단=====')
  for j in range(1, 10) :
    # result = i * j
    # print(f'{i} X {j} = {"%2d" % result}')
    print(f'{i:2} X {j:2} = {i * j :2}') # 2자리씩 주게하는 방법
```

```python
# 문제) 단수를 입력받아 해당 구구단을 출력하시오.

dansu = int(input('단수를 입력하세요 : '))
# print(type(dansu)) # String 타입

for num in range(1, 9+1) :
  print(f'{dansu : 2} X {num : 2} = {dansu * num :>2}')
```

```python
def sum_all(*args):
  print(type(args))

  for ele in args:
    print(ele)

sum_all(1, 2, 3)
```

```python
# 가변인자 :   *:튜플, **:딕셔너리
def sum_all2(*args1, **args2):
  print(type(args1))
  print(type(args2))

  for ele in args1:
    print(ele)
  print('='*20)

  for key in args2.keys():
    print(key)
  print('='*20)
  for key, val in args2.items():
    print(key, val)

print('★★★★★★Case1)호출★★★★★★')
sum_all2(1, 2, 3, name='홍길동', age=10, gender='남자')

x = (1, 2, 3)                                     # 튜플
y = {'name':'홍길동', 'age':10, 'gender':'남자'}  # 딕셔너리
print('★★★★★★Case2)호출★★★★★★')
sum_all2(*x, **y)   # 이렇게 호출해도 됨
```

```python
g_var = 100     #전역변수 : 함수 밖에 선언 된 변수

def f(p_var=300):        #매개변수 : 함수내에서 지역변수 처럼 사용 됨
  l_var = 200   #로컬변수 : 함수 내에 선언 된 변수
  global g_var

  print(g_var) # 100
  g_var = 400

  print(g_var) # 400

  print(p_var) # 300

  print(l_var) # 200


f()
```

```python
list = [1, 2, 3]
list2 = []
for ele in list :
  list2.append(ele*2)

print(list2)

print('='*20)

def mul(x) :
  return x * 2

result = map(lambda x : x*2, list2)
for ele in result :
  print(ele)

print('='*20)

result2 = map(mul, list2)
for ele in result2 :
  print(ele)
```
