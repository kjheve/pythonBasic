#리스트 연습

```python
a = [1, 2, 3, ['a', 'b', 'c']]
print(a[0])
print(a[-1])
print(a[-1][1])
print(a[2])

# 빈 배열만들기
b = []
c = list()
print (b, c)

#요소 추가 및 제거
b.append(1)
b.append(2)
b.append(3)
print(b)
b.remove(2)
print(b)
```

```python
print(a)
print('-----------------------------------')
# 슬라이싱 : 범위 선택자 리스트[시작 : 종료(미포함) : 스텝]

print (a[1:])
print (a[1:3])
print (a[:]) #전체 요소
print (a[1::2])
```

### 튜플

```python
a = (1, 2, 3)
print (type(a))
print(a[0])
print(a[-1])
#a[1] = 10 #요소 변경 안됨
#del a[1] #요소 삭제 안됨

#튜플을 리스트로 변환
b = list(a)
print(type(b))
b[1] = 10
print(b)

#리스트를 튜플로 변환
c = tuple(b)
print(type(c))
#요소 갯수 확인
print(len(c))
```
