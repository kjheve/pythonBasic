## 필요 없는 컬럼 삭제하기

```python
df = df.drop(['시설명','시군구','위도','경도','도로명주소','전화번호','홈페이지','휴무일','운영시간','지번주소','주차가능여부','이용료정보','크기제한','기타제한사항','장소요약','동반시추가요금'],axis=1)
df
```

## 컬럼명 변경 

```python
# 변경 안해도 되는데 연습용이로 바꿈!

df = df.replace(['카페및식당'], '카페 및 식당')
df = df.replace(['반려동물용품'], '반려용품')
df = df.replace(['문화'], '문화여행시설')
df = df.replace(['위탁및숙박'], '위탁 및 숙박')
df = df.replace(['미용'], '미용시설')
df['시설분류'].unique()
```

```python
df.head()
```

## 멀티인덱스로 변경

```python
# 행정구역별 시설의 수 확인
df.groupby('행정구역')['시설분류'].value_counts()
```

```python
# 멀티인덱스로 변경된 데이터는 이제 df2에 
df2 = df.groupby('행정구역')['시설분류'].value_counts()
df2 
```

## 멀티인덱스를 컬럼으로 unstack

```python
# ★ 멀티인덱스의 '시설분류'를 컬럼으로 변경
df2 = df2.unstack()
df2
```

## value 값을 수량 -> 비율(%) 로 변경

```python
#각 행정구역별 시설분류의 값을 비율(%)로

df2_percentage = df2.div(df2.sum(axis=1), axis=0) * 100 
df2_percentage
```

## 비율의 '합계' 컬럼 생성

```python
# 모든 시도별 시설분류의 비율의 합이 100%인지 확인하기 위해 '합계' 컬럼생성

df2_percentage['합계']= df2_percentage['동물병원'] + df2_percentage['동물약국'] + df2_percentage['문화여행시설'] + df2_percentage['미용시설'] + df2_percentage['반려용품'] + df2_percentage['위탁 및 숙박'] + df2_percentage['카페 및 식당']
df2_percentage
```

```python
# value값이 int -> float 실수 값으로 변경됨
df2_percentage.info() 
```

## value값을 소수점 첫째자리에서 반올림처리

```python
# 깔끔하게 소수점 첫째자리에서 반올림
df2 = df2_percentage.round(1)
df2
```

```python
df2.info()
```

## 확인용 '합계' 컬럼 삭제

```python
df2 = df2.drop('합계', axis=1)
df2
```

```python
df2
```

## 시각화 (누적 파이 차트)
### 가로형 누적 막대 차트

```python
df2.plot(kind='barh', 
        stacked=True, # 누적으로 그래프 쌓을 것이다
        colormap='tab10', # 컬러 설정

        figsize=(12, 8)) # 차트 사이즈

plt.legend( ncol=7, loc=(-0.2, -0.2)) # 범례 디자인: 가로로 7개 나열, 위치는 차트 기준 하단(마이너스)
plt.xlabel("비율(%)") # 가로축 라벨
plt.ylabel("행정구역") # 세로축 라벨

# plt.box(False) # 테두리 사각박스없애기

plt.show()
```

### 가로형 누적 막대 차트+ 비율 텍스트 추가

```python
ax = df2.plot(kind='barh', 
              stacked=True, 
              colormap='Set2', # 색상은 다른 것 아무거나로 변경 가능! 
              width=0.7, # 막대 넓이 ('1'이 최대값)
              figsize=(15, 9))


#bbox_to_anchor의 첫 번째 값은 좌우 위치(0.5는 중앙), 두 번째 값은 상하 위치.
#상하 위치 값이 1보다 크면 그래프 위로 범례가 이동
plt.legend( ncol=7, loc='upper center', bbox_to_anchor=(0.49, 1.05))
plt.xlabel("비율(%)")
plt.ylabel("행정구역")

# 비율 text 작성 (소수점 첫째자리까지)
for p in ax.patches:
    width = p.get_width()
    if width >= 2.0:  # 수치가 2% 이상인 경우에만 텍스트를 나타내기
        rounded_width = round(width, 1)  # 소수점 첫째자리까지만 나타내기
        plt.text(p.get_x() + width/2.,   # 막대 x의 중앙 좌표
                 p.get_y() + p.get_height()/2., # 막대 y의 중앙 좌표
                 f'{rounded_width}', # 나타낼 숫자
                 
                 #ha='center': 텍스트를 수평 방향으로 중앙에 정렬합니다.
                 #va='center': 텍스트를 수직 방향으로 중앙에 정렬합니다
                 ha='center', va='center',   
                 
                 color='white', # 폰트 색상
                 fontweight='bold', # 폰트 굵기
                 fontsize=12)  # 폰트 사이즈

plt.box(False)    
plt.show()
```
