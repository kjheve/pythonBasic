### merge()

```python
# KEY 컬럼명이 같은 경우
left = pd.DataFrame({
    'KEY' : ['K0', 'K1', 'k2', 'K3'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
})
right = pd.DataFrame({
    'KEY' : ['K0', 'K1', 'K3', 'k4'],
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
})
```

```python
display(left)
display(right)
```

```python
result = pd.merge(left, right, on='KEY')
result
```

```python
pd.merge(left, right, on='KEY', how='left')
```

```python
pd.merge(left, right, on='KEY', how='right')
```

```python
pd.merge(left, right, on='KEY', how='outer')
```

```python
# KEY 컬럼명이 다른 경우
left = pd.DataFrame({
    'KEY_LEFT' : ['K0', 'K1', 'k2', 'K3'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
})
right = pd.DataFrame({
    'KEY_RIGHT' : ['K0', 'K1', 'K3', 'k4'],
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
})
display(left)
display(right)
```

```python
pd.merge(left, right, left_on='KEY_LEFT', right_on='KEY_RIGHT')
```

```python
pd.merge(left, right, left_on='KEY_LEFT', right_on='KEY_RIGHT', how='left')
```

```python
pd.merge(left, right, left_on='KEY_LEFT', right_on='KEY_RIGHT', how='right')
```

```python
pd.merge(left, right, left_on='KEY_LEFT', right_on='KEY_RIGHT', how='outer')
```

```python
# left 이렇게 해도 결과는 동일하다
left.merge(right, left_on='KEY_LEFT', right_on='KEY_RIGHT')
```

### join()

```python
# join
# KEY 컬럼명이 같은 경우
left = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
}, index=['K0', 'K1', 'k2', 'K3'])
right = pd.DataFrame({
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
}, index=['K0', 'K1', 'K3', 'k4'])
```

```python
display(left)
display(right)
```

```python
# merge는 기본이 inner
# join은 기본이 left outer join
print('--left--')
display(left.join(right))
print('--inner--')
display(left.join(right, how='inner'))
print('--right--')
display(left.join(right, how='right'))
print('--full outer--')
display(left.join(right, how='outer'))
```
