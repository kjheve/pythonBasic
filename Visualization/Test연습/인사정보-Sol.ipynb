### (상) 각 직급별 연봉의 최대값, 최소값, 평균값을 계산하세요.

```python
df_11 = df.groupby('직급')[['이름','연봉(만원)']].max()
df_11.columns = ['이름', '최대값'] # 컬럼 이름 변경
min = df.groupby('직급')['연봉(만원)'].min()
mean = df.groupby('직급')['연봉(만원)'].mean()
df_11['최소값'] = min # 컬럼 추가
df_11['평균값'] = mean # 컬럼 추가
df_11
```

***
* sol

```python
df.groupby('직급')['연봉(만원)'].agg(['max', 'min', 'mean'])
```

###  (상) 연봉의 범위(최소값과 최대값 사이의 차이)가 가장 큰 부서를 찾으세요.

```python
df_15_col = df_11['최대값'] - df_11['최소값']
```

```python
df_11['연봉의범위'] = df_15_col
```

```python
df_11
```

```python
max = df_11['연봉의범위'].max()
filt = df_11['연봉의범위'] == max
df_11[filt]
```
