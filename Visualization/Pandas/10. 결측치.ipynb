## 데이터 채우기 fillna

```python
df.fillna('없음') # NaN 데이터를 빈 칸으로 채움
```

```python
import numpy as np
```

```python
df['학교'] = np.nan # 학교 데이터 전체를 NaN으로 채움
df
```

```python
df.fillna('모름', inplace=True)
df
```

```python
# SW특기 데이터 중에서 NaN에 대해서 채움
df['SW특기'].fillna('확인 중', inplace=True)
df
```

### 데이터 제외하기 dropna

```python
# 전체 데이터 중에서 NaN을 포함하는 데이터 삭제
df.dropna()
```

- axis : index(0) or columns(1)
- how : any or all

```python
# Nan가 하나라도 있는 row 삭제
df.dropna(axis='index', how='any')
```

```python
# NaN이 하나라도 있는 coulmn 삭제
df.dropna(axis='columns')
```

```python
df['학교'] = np.nan
```

```python
# 데이터 전체가 NaN 인 경우에만 Column 삭제
df.dropna(axis='columns', how='all')
```
