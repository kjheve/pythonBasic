### 형태소 분석 함수 정의

```python
def extract_korean_from_series(series):
    
    # 정규 표현식을 사용하여 한글만 추출 (함수 안에 함수를 정의)
    def extract_korean(text):
        # 한글을 제외한 모든 문자 제거
        text = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣\s]',' ',text)
        # 공백 2개 이상인 경우 1개로 변환
        text = re.sub(r'\s+',' ',text)
        return text

    # 시리즈의 각 요소에 대해 한글 추출 함수 적용
    return series.map(extract_korean)
```

```python
# 소요시간 추가
def extract_adjectives_from_series(pecab, series):

    adjectives = set()  # 중복을 제거하기 위해 집합 사용
    # 시리즈의 각 요소에 대해 반복
    for i, text in enumerate(series):
        
        start = time.time()
        
        pecab_pos = pecab.pos(text)
        join_text = []
        for text,pos in pecab_pos:
            if pos == 'VA' or pos == 'NNP':
#             if not pos.startswith('J'): # 조사만 제거
                join_text.append(f'{text} - {pos}')
    
        adjectives.update(set(join_text))
        print(f'{i+1}번째 수행../ 소요시간 : {time.time() - start}')  
    return list(adjectives)
```

#### okt (up)

```python
# 시리즈에서 중복을 제거한 형용사와 동사 추출 후 리스트로 반환
def extract_adjectives_from_series_okt(analyzer, series):
    
    words = set()  # 중복을 제거하기 위해 집합 사용

    # 시리즈의 각 요소에 대해 반복
    for i, text in enumerate(series):
        start = time.time()
        analyzer_pos = analyzer.pos(text)
        for word, pos in analyzer_pos:
            
            if len(word) > 1 and (pos == 'Verb' or pos == 'Adjective'):
                words.add((pos, word))

#         print(f'{i+1}번째 수행.../ 소요시간 : {time.time() - start}')

    # (품사, 단어) 형식의 튜플을 리스트로 변경하여 반환
    return [[pos, word] for pos, word in words]
```

#### kkma()

```python
def extract_adjectives_from_series_kkma(kkma, series):

    adjectives = set()  # 중복을 제거하기 위해 집합 사용
    # 시리즈의 각 요소에 대해 반복
    for i, text in enumerate(series):
        
        kkma_pos = kkma.pos(text)
        join_text = []
        for text,pos in kkma_pos:
            if pos == 'VA':
#             or pos == 'NNG':  # 형용사와 보통명사 선택
                join_text.append(f'{text} - {pos}')
        adjectives.update(set(join_text))
    
    return list(adjectives)
```

## 감성사전 만들기

```python
verb_adjective_1t,verb_adjective_2t,verb_adjective_3t
```

```python
combined_array = np.vstack((verb_adjective_1t,verb_adjective_2t,verb_adjective_3t))
combined_array
```

```python
sentiment_dict = pd.DataFrame(combined_array,columns=['품사','단어'])
sentiment_dict = sentiment_dict.sort_values(by=['품사','단어'], ascending=[True,True])
```
