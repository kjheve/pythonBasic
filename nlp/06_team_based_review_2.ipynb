## 정규화

```python
from konlpy.tag import Okt
```

```python
def extract_norm(words):
    if pd.isna(words):
        return words
    
    okt = Okt()
    norm = []
    for word in words:
        # 정규화
        norm = okt.normalize(words)
    return norm
```

## 어간 추출

```python
# 어간 추출
def extract_stem(words):
    if pd.isna(words):
        return words
    
    okt = Okt()
    stems = []
    for word in words:
        # 어간 추출
        stems = okt.morphs(words,stem=True)
    return stems
```

```python
from tqdm import tqdm
tqdm.pandas()
```

## 함수적용
